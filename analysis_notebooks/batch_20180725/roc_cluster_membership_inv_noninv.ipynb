# Objective
Investigate TPR and FPR using inverted and non-inverted images.

Variable along curve: number of unique workers required for a cluster to be "correct."

```python
from SpotAnnotationAnalysis import SpotAnnotationAnalysis
from BaseAnnotation import BaseAnnotation
from QuantiusAnnotation import QuantiusAnnotation

worker_marker_size = 8
cluster_marker_size = 40
bigger_window_size = False
img_height = 300
correctness_threshold = 4
clustering_params = ['AffinityPropagation', -350]
```

# Results

For all of the images in this set (except 50 spots, SNR_mu = 10), the curves for the inverted and original images are very similar. Inverting would not have been a useful part of the pipeline to prepare these six images (cell background; num_spots = 50, 100, 150; SNR_mu = 5, 10).

# Background: Cells

```python
bg_type = 'cells'
```

## 50 spots, SNR = 5, 10

```python
img_names = ['MAX_C3-ISP_300_1_nspots50_spot_sig1.75_snr5_2.5',
             'MAX_C3-ISP_300_1_nspots50_spot_sig1.75_snr10_2.5']

for img_name in img_names:
    plot_title = img_name
    
    gen_date = '20180725'
    json_filename = 'InvTest.json'
    img_filename = img_name+'spot_img_inv.png'
    img_filepath = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/spot_images/'+bg_type+'/'+img_name+'spot_img_inv.png'
    csv_filepath_1 = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/spot_data/'+bg_type+'/'+img_name+'_coord_snr_list.csv'
    json_filepath = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/'+json_filename
    ba = QuantiusAnnotation(json_filepath)
    sa = SpotAnnotationAnalysis(ba)
    anno_all = ba.df()
    anno_one_snr_1 = ba.slice_by_image(anno_all, img_filename)
    
    gen_date = '20180719'
    json_filename = 'SynthData_cells.json'
    img_filename = img_name+'spot_img.png'
    img_filepath = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/spot_images/'+bg_type+'/'+img_name+'spot_img.png'
    csv_filepath_2 = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/spot_data/'+bg_type+'/'+img_name+'_coord_snr_list.csv'
    json_filepath = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/'+json_filename
    ba = QuantiusAnnotation(json_filepath)
    sa = SpotAnnotationAnalysis(ba)
    anno_all = ba.df()
    anno_one_snr_2 = ba.slice_by_image(anno_all, img_filename)
    
    sa.plot_cluster_membership_threshold_roc(anno_one_snr_1, anno_one_snr_2, clustering_params, csv_filepath_1, csv_filepath_2, img_height, correctness_threshold, plot_title, bigger_window_size)
```
