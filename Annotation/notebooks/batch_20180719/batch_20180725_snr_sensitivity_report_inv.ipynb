# Objective
Investigate the sensitivity of the workers to SNR using inverted and non-inverted images.
For a given dataset, take the subset of reference spots with an SNR > n and calculate the fraction of that subset that were detected by the turkers. Build a curve by varying n (SNR_min through SNR_max).

```python
from SpotAnnotationAnalysis import SpotAnnotationAnalysis
from BaseAnnotation import BaseAnnotation
from QuantiusAnnotation import QuantiusAnnotation

worker_marker_size = 8
cluster_marker_size = 40
bigger_window_size = False
img_height = 300
correctness_threshold = 4
clustering_params = ['AffinityPropagation', -350]
```

# Results

Min SNR for 100% spot detection w/ original images => min SNR for 100% spot detection w/ inverted images. Visually, the difference is not dramatic. Inverting the image seems to be more helpful when mean SNR = 5 than when mean SNR = 10.

The shapes of the curves for the inverted and original images actually look very similar in some cases. Side-by-side comparison here: https://docs.google.com/document/d/1gwL3vfoalTm_riL_1STyQnFJrXG0J9qU4pKOHQj4rx0/edit?usp=sharing

| Background | # Spots | Mean SNR | Min SNR for 100% detection | Min SNR for 100% detection, inv'd image |
| --- | --- | --- |--- | --- |
| cells | 50 | 5 | 10 | 6 |
| cells | 50 | 10 | 12 | 12|
| cells | 100 | 5 | 8 | 8 |
| cells | 100 | 10 | 15? | 14 |
| cells | 150 | 5 | 12 | 10 |
| cells | 150 | 10 | 14 | 13 |

**Takeaway:** When the SNR is low (closer to 5), inverting the image can help increase yield. When the SNR is high (closer to 10), inverting the image may be an unecessary step.

## 50 spots, SNR = 5, 10

```python
img_names = ['MAX_C3-ISP_300_1_nspots50_spot_sig1.75_snr5_2.5',
             'MAX_C3-ISP_300_1_nspots50_spot_sig1.75_snr10_2.5']

for img_name in img_names:
    img_filename = img_name+'spot_img_inv.png'
    img_filepath = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/spot_images/'+bg_type+'/'+img_name+'spot_img_inv.png'
    csv_filepath = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/spot_data/'+bg_type+'/'+img_name+'_coord_snr_list.csv'
    json_filepath = '/Users/jenny.vo-phamhi/Documents/FISH-annotation/Annotation/gen_'+gen_date+'/'+json_filename
    ba = QuantiusAnnotation(json_filepath)
    sa = SpotAnnotationAnalysis(ba)
    anno_all = ba.df()
    anno_one_snr = ba.slice_by_image(anno_all, img_filename)
    plot_title = img_name
    sa.plot_snr_sensitivity(anno_one_snr, clustering_params, csv_filepath, img_height, img_filename, correctness_threshold, plot_title, bigger_window_size)
```
